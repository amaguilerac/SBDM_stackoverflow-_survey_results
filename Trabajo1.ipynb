```
#Cargar datos
import pandas as pd
from sodapy import Socrata

client = Socrata("www.datos.gov.co", None)

results = client.get("gt2j-8ykr",limit= 200000000)

results_df = pd.DataFrame.from_records(results)
results_df
```

```
#Elegir columnas

covid = results_df[['id_de_caso','fecha_de_notificaci_n','c_digo_divipola','ciudad_de_ubicaci_n','departamento','atenci_n','edad','sexo','tipo','estado','pa_s_de_procedencia','fis','fecha_diagnostico','fecha_recuperado','fecha_reporte_web','tipo_recuperaci_n','codigo_departamento','codigo_pais','pertenencia_etnica','nombre_grupo_etnico','fecha_de_muerte']].copy()
covid
```

```
# Cambio nombres
covid = covid.rename(columns={'id_de_caso':'ID','fecha_de_notificaci_n':'fecha_notificacion','c_digo_divipola':'cod D','ciudad_de_ubicaci_n':'ciudad_ubicacion', 'atenci_n': 'atencion', 'pa_s_de_procedencia': 'pais_procedencia','fecha_diagnostico': 'fecha_diagnostico','fecha_reporte_web':'fecha_reporte','tipo_recuperaci_n':'tipo_recuperacion','nombre_grupo_etnico':'grupo_etnico','fecha_de_muerte':'fecha_muerte'})
covid
```

```
#Cambio formato fecha
covid["fecha_notificacion"]= covid["fecha_notificacion"].str.split("T",2).str[0]
covid["fis"]= covid["fis"].str.split("T",2).str[0]
covid["fecha_diagnostico"]= covid["fecha_diagnostico"].str.split("T",2).str[0]
covid["fecha_recuperado"]= covid["fecha_recuperado"].str.split("T",2).str[0]
covid["fecha_reporte"]= covid["fecha_reporte"].str.split("T",2).str[0]
covid
```

```
#Tabla de frecuencias
Frecuencias=pd.DataFrame(covid.groupby('ciudad_ubicacion')['ciudad_ubicacion'].count())
Frecuencias.columns=['conteo']
Frecuencias=Frecuencias.reset_index()
Frecuencias['prob']=Frecuencias['conteo']/Frecuencias['conteo'].sum()
Frecuencias
```
